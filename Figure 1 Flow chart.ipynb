## Code for the Figure 1

```python
with schemdraw.Drawing() as d:
#    d.config(fontsize=14)
    
    #Steps
    d += (b := flow.Start(w=8.5, h=2).label('New or established model in\n freshwater ecology', fontsize=16))
    d += flow.Arrow().down(d.unit/2)
    
    d += (step1 := flow.Box(w=9, h=2).label('1) Carefully read and interpret the\n postulates of the new model')
          .fill('white'))
    d += flow.Arrow().down(d.unit/2)
    
    d += (step2 := flow.Box(w=9, h=2).label('2) Carefully examine the mechanisms\n and methodology of the new model')
          .fill('white'))
    d += flow.Arrow().down(d.unit/2)
    
    d += (step3 := flow.Box(w=9, h=2).label('3) Understand the circumstances in\n which the model was proposed')
          .fill('white'))    
    d += flow.Arrow().down(d.unit/2)
    
    d += (step4 := flow.Box(w=9, h=2).label('4) Gather and analyze data following\n adjusted methodologies')
          .fill('white'))    
    d += flow.Arrow().down(d.unit/2)
    
    d += (step5 := flow.Box(w=9, h=2).label('5) The results are consistent or\n inconsistent with the original model')
          .fill('white'))
    d += flow.Arrow().down(d.unit/2)
    
    d += (step6 := flow.Box(w=9, h=2).label('6) Original postulates are empirically\n supported by the results, but the\n predictive power of the model is limited')
          .fill('white'))
    d += flow.Arrow().down(d.unit/2)
    
    d += (step7 := flow.Box(w=9, h=2).label('7) Make recommendations').fill('white'))


    # Activities
    d += flow.Arrow().right(d.unit/2).at(step1.E)
    d += (act1 := flow.Box(w=8, h=2).label('Understand your local context (both\n socioeconomic and ecological)')
          .fill('white'))
    
    d += flow.Arrow().right(d.unit/2).at(step2.E)
    d += (act2 := flow.Box(w=8, h=2).label('Make necessary adjustments')
         .fill('white'))

    d += flow.Arrow().right(d.unit/2).at(step3.E)
    d += (act3 := flow.Box(w=8, h=2).label('Evaluate model application within\n your local context').
          fill('white'))
 
    d += flow.Arrow().right(d.unit/2).at(step4.E)
    d += (act4 := flow.Box(w=8, h=2).label('Ensure you have enough data to\n conduct rigorous analyses and\n properly test model postulates')
          .fill('white'))
    
    d += flow.Arrow().right(d.unit/2).at(step5.E)
    d += (act5 := flow.Box(w=8, h=2).label('Return to the original postulates')
          .fill('white'))
    
    d += flow.Arrow().right(d.unit/2).at(step6.E)
    d += (act6 := flow.Box(w=8, h=2).label('Determine limitations and\n revise applicability of the model')
          .fill('white'))

    d += flow.Arrow().right(d.unit/2).at(step7.E)
    d += (act7 := flow.Box(w=8, h=2).anchor('W').label('Provide appropriate feedback\n based on local experience')
         .fill('white'))
    
    d += flow.Arrow().down(d.unit/2).at(act7.S)
    d += flow.Start(w=8.5, h=2).label('New or adapted model\n for Latinamerican context', fontsize=16)
    
    
    
    #Phase names.  Frame each phase. 
    d += (phase1 := elm.EncircleBox([step1, step2, step3, act1, act2, act3], padx=.85,pady=0.5).linestyle('--')
          .linewidth(3).color('#238b45').fill('#ccece6').label(r'$\bfGroundwork$', loc='left', rotate=90, fontsize=17))
    
    d += (phase2 := elm.EncircleBox([step4, act4], padx=.85,pady=0.5).linestyle('--')
          .linewidth(3).color('#88419d').fill("#dadaeb").label(r'$\bfImplementation$', loc='left',rotate=90, fontsize=17))
        
    d += (phase2 := elm.EncircleBox([step5, step6, act5, act6], padx=.85,pady=0.5).linestyle('--')
          .linewidth(3).color('#2171b5').fill("#c6dbef").label(r'$\bfValidation$', loc='left',rotate=90, fontsize=17))
    
    d += (phase3 := elm.EncircleBox([step7, act7], padx=0.85, pady=.5).linestyle('--')
          .linewidth(3).color('#cb181d').fill('#fee5d9').label(r'$\bfAdaptation$', loc='left', rotate=90, fontsize=17))
    
    
    # Save and export figure
    d.save("Figure 1.jpg", dpi=300)
    d.save("Figure 1.eps", dpi=300)
        
```
